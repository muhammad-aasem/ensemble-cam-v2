# Dataset Preparation for NIH Chest X-ray Dataset

This notebook provides a comprehensive dataset preparation workflow for the NIH Chest X-ray dataset.

## Overview

The workflow includes:
- Downloading the dataset from Hugging Face
- Extracting zip files
- Creating object detection subset (BBox)
- Creating classification dataset
- Excluding overlapping images
- Preprocessing images (resize, format conversion)
- Creating a small dataset for development
- Optional cleanup of downloaded files

## Prerequisites

Make sure you have:
- Python 3.13.2+ with UV package manager
- Sufficient disk space (~50GB for full dataset)
- Hugging Face access token configured

---

## Workflow Steps

1. **Download Dataset** - Download from Hugging Face
2. **Extract Files** - Extract all zip files
3. **Create BBox Subset** - Object detection subset
4. **Create Classification Dataset** - Disease classification
5. **Exclude Overlapping Images** - Remove BBox images from classification
6. **Preprocess Images** - Resize and format conversion
7. **Create Small Dataset** - Development subset
8. **Cleanup** - Remove original downloaded files (optional)

After completing this notebook, you can proceed to `playbook2_train_Xception.ipynb` for model training.

````python
import os
import sys
import subprocess
from pathlib import Path
import time

# Add project root to path
project_root = Path.cwd().parent
sys.path.append(str(project_root))

print(f"Project root: {project_root}")
print(f"Current working directory: {os.getcwd()}")
print(f"Dataset preparation starting...")
````

## Step 1: Dataset Download

Download the NIH Chest X-ray dataset from Hugging Face.


````python
print("Step 1: Downloading NIH Chest X-ray dataset...")
print("=" * 50)

# Run dataset download script
result = subprocess.run([
    "uv", "run", "python", 
    str(project_root / "src" / "NIH-Chest-X-ray" / "dataset_1_download.py")
], capture_output=True, text=True)

print("STDOUT:")
print(result.stdout)
if result.stderr:
    print("STDERR:")
    print(result.stderr)

if result.returncode == 0:
    print("Dataset download completed successfully!")
else:
    print(f"Dataset download failed with return code: {result.returncode}")
    print("Please check the error messages above.")
````

## Dataset Preparation Summary

Let's check what datasets we have created:


````python
print("Dataset Preparation Summary")
print("=" * 50)

dataset_dir = project_root / "dataset" / "NIH-Chest-X-ray"

if dataset_dir.exists():
    print(f"Dataset directory: {dataset_dir}")
    print("\nCreated datasets:")
    
    for subdir in dataset_dir.iterdir():
        if subdir.is_dir():
            # Count files in subdirectory
            file_count = len(list(subdir.rglob("*.jpg"))) + len(list(subdir.rglob("*.png")))
            print(f"  - {subdir.name}: {file_count:,} images")
            
            # If it's a classification dataset, show class breakdown
            if "Classification" in subdir.name:
                class_dirs = [d for d in subdir.iterdir() if d.is_dir()]
                print(f"    Classes: {len(class_dirs)}")
                for class_dir in sorted(class_dirs)[:5]:  # Show first 5 classes
                    class_files = len(list(class_dir.glob("*.jpg")))
                    print(f"      {class_dir.name}: {class_files:,} images")
                if len(class_dirs) > 5:
                    print(f"      ... and {len(class_dirs) - 5} more classes")
else:
    print("Dataset directory not found. Please run the dataset preparation steps first.")
````
